```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate realistic dummy data: actual vs budgeted revenue for multiple months
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
actual_revenue = np.random.randint(12000, 18000, size=12)
budgeted_revenue = actual_revenue + np.random.randint(-1500, 1500, size=12)

# Create DataFrame
df_variance = pd.DataFrame({
    'Month': months,
    'Actual Revenue': actual_revenue,
    'Budgeted Revenue': budgeted_revenue
})

# Calculate variance
df_variance['Variance'] = df_variance['Actual Revenue'] - df_variance['Budgeted Revenue']

# Set Seaborn theme
sns.set_theme(style="whitegrid")

# Assign colors based on positive or negative variance
df_variance['Color'] = df_variance['Variance'].apply(lambda x: 'green' if x >= 0 else 'red')

# Create the plot
plt.figure(figsize=(14, 8))
sns.barplot(
    data=df_variance,
    x='Month',
    y='Variance',
    hue='Color',
    dodge=False,
    palette={'green': 'green', 'red': 'red'},
    legend=False
)

# Enhance visualization
for i, v in enumerate(df_variance['Variance']):
    plt.text(i, v + (500 if v >= 0 else -1000), f"{v:.0f}", ha='center', va='center', fontsize=10, color='black')

plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.title('Variance Analysis: Actual vs Budgeted Revenue', fontsize=20, fontweight='bold')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Variance ($)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Dummy data: Base values
base_revenue = 100000
base_cost = 60000

# Define scenarios with different revenue growth rates
scenarios = {
    'Scenario': ['Best Case', 'Base Case', 'Worst Case'],
    'Revenue Growth (%)': [20, 10, -5]
}

# Create DataFrame
df_scenarios = pd.DataFrame(scenarios)

# Calculate projected revenue and profit for each scenario
df_scenarios['Projected Revenue'] = base_revenue * (1 + df_scenarios['Revenue Growth (%)'] / 100)
df_scenarios['Projected Profit'] = df_scenarios['Projected Revenue'] - base_cost
df_scenarios['Profit Margin (%)'] = (df_scenarios['Projected Profit'] / df_scenarios['Projected Revenue']) * 100

# Set Seaborn theme
sns.set_theme(style="whitegrid")

# Create the plot
plt.figure(figsize=(12, 8))
sns.barplot(
    data=df_scenarios,
    x='Scenario',
    y='Projected Profit',
    palette='viridis'
)

# Add labels to each bar
for i, v in enumerate(df_scenarios['Projected Profit']):
    plt.text(i, v + 2000, f"${v:,.0f}", ha='center', fontsize=12)

# Enhance visualization
plt.title('Scenario Analysis: Profitability Under Different Scenarios', fontsize=18, fontweight='bold')
plt.xlabel('Scenario', fontsize=14)
plt.ylabel('Projected Profit ($)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dummy data: Revenue by product line
data = {
    'Product Line': ['A', 'B', 'C', 'D', 'E'],
    'Last Month Revenue': [20000, 30000, 25000, 40000, 15000],
    'This Month Revenue': [15000, 25000, 20000, 30000, 12000]
}

# Create DataFrame
df_root_cause = pd.DataFrame(data)

# Calculate revenue drop
df_root_cause['Revenue Drop'] = df_root_cause['Last Month Revenue'] - df_root_cause['This Month Revenue']

# Identify root cause (largest drop)
df_root_cause['Is Root Cause'] = df_root_cause['Revenue Drop'] == df_root_cause['Revenue Drop'].max()

# Set Seaborn theme
sns.set_theme(style="whitegrid")

# Create the plot
plt.figure(figsize=(12, 8))
sns.barplot(
    data=df_root_cause,
    x='Product Line',
    y='Revenue Drop',
    hue='Is Root Cause',
    dodge=False,
    palette={True: 'red', False: 'blue'},
    legend=False
)

# Add labels to each bar
for i, v in enumerate(df_root_cause['Revenue Drop']):
    plt.text(i, v + 1000, f"{v:.0f}", ha='center', fontsize=12, color='black')

# Highlight the root cause in the title
root_cause_line = df_root_cause.loc[df_root_cause['Is Root Cause'], 'Product Line'].values[0]
plt.title(f"Root Cause Analysis: Revenue Drop by Product Line\nLargest Drop: {root_cause_line}",
          fontsize=18, fontweight='bold')
plt.xlabel('Product Line', fontsize=14)
plt.ylabel('Revenue Drop ($)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Dummy data: Marketing spend vs. sales
np.random.seed(42)  # For reproducibility
marketing_spend = np.random.randint(5000, 20000, size=20)
sales = marketing_spend * np.random.uniform(1.5, 2.0, size=20) + np.random.randint(-5000, 5000, size=20)

# Create DataFrame
df_correlation = pd.DataFrame({
    'Marketing Spend ($)': marketing_spend,
    'Sales ($)': sales
})

# Calculate correlation
correlation_value = df_correlation.corr().iloc[0, 1]

# Set Seaborn theme
sns.set_theme(style="whitegrid")

# Create the scatter plot with regression line
plt.figure(figsize=(12, 8))
sns.regplot(
    data=df_correlation,
    x='Marketing Spend ($)',
    y='Sales ($)',
    scatter_kws={'s': 100, 'alpha': 0.6},
    line_kws={'color': 'red', 'alpha': 0.8, 'linewidth': 2}
)

# Add correlation value as a text annotation
plt.text(
    x=df_correlation['Marketing Spend ($)'].max() * 0.6,
    y=df_correlation['Sales ($)'].max() * 0.9,
    s=f"Correlation: {correlation_value:.2f}",
    fontsize=14,
    color='black',
    bbox=dict(facecolor='white', edgecolor='gray', boxstyle='round,pad=0.5')
)

# Enhance visualization
plt.title('Correlation Analysis: Marketing Spend vs Sales', fontsize=18, fontweight='bold')
plt.xlabel('Marketing Spend ($)', fontsize=14)
plt.ylabel('Sales ($)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Dummy data: Monthly revenue
np.random.seed(42)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
revenue = np.random.randint(10000, 20000, size=12)
# Introduce outliers
revenue[2] = 35000  # March
revenue[8] = 5000   # September

# Create DataFrame
df_outliers = pd.DataFrame({
    'Month': months,
    'Revenue': revenue
})

# Set Seaborn theme
sns.set_theme(style="whitegrid")

# Create the boxplot to detect outliers
plt.figure(figsize=(12, 8))
sns.boxplot(data=df_outliers, x='Revenue', color='skyblue')

# Add annotations for outliers
outliers = df_outliers[
    (df_outliers['Revenue'] < df_outliers['Revenue'].quantile(0.25) - 1.5 * (df_outliers['Revenue'].quantile(0.75) - df_outliers['Revenue'].quantile(0.25))) |
    (df_outliers['Revenue'] > df_outliers['Revenue'].quantile(0.75) + 1.5 * (df_outliers['Revenue'].quantile(0.75) - df_outliers['Revenue'].quantile(0.25)))
]

for i, row in outliers.iterrows():
    plt.text(row['Revenue'], 0.15, f"{row['Month']}\n${row['Revenue']}", fontsize=10, color='red')

# Enhance the visualization
plt.title('Outliers Detection in Monthly Revenue', fontsize=18, fontweight='bold')
plt.xlabel('Revenue ($)', fontsize=14)
plt.yticks([])  # Hide y-axis
plt.tight_layout()

# Show the plot
plt.show()

# Print outlier details
print("Detected Outliers:")
print(outliers)
```

```python
# Dummy data: Historical monthly revenue
np.random.seed(42)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
revenue = np.random.randint(10000, 20000, size=12)

# Create DataFrame
df_forecasting = pd.DataFrame({
    'Month': months,
    'Revenue ($)': revenue
})

# Calculate a 3-month moving average for forecasting
df_forecasting['3-Month Moving Average ($)'] = df_forecasting['Revenue ($)'].rolling(window=3).mean()

# Extend dataset to forecast next 3 months
future_months = ['Jan (Next Year)', 'Feb (Next Year)', 'Mar (Next Year)']
forecast_values = df_forecasting['3-Month Moving Average ($)'][-3:].mean()  # Simple mean for next months
future_data = pd.DataFrame({
    'Month': future_months,
    'Revenue ($)': [None] * 3,
    '3-Month Moving Average ($)': [forecast_values] * 3
})
df_forecasting = pd.concat([df_forecasting, future_data], ignore_index=True)

# Set Seaborn theme
sns.set_theme(style="whitegrid")

# Plot historical and forecasted revenue
plt.figure(figsize=(14, 8))
sns.lineplot(data=df_forecasting, x='Month', y='Revenue ($)', label='Actual Revenue', marker='o')
sns.lineplot(data=df_forecasting, x='Month', y='3-Month Moving Average ($)', label='Forecast', linestyle='--', color='orange')

# Enhance visualization
plt.title('Revenue Forecasting Using Moving Average', fontsize=18, fontweight='bold')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Revenue ($)', fontsize=14)
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Dummy data: Customer revenue
np.random.seed(42)
data = {
    'Customer': [f'Cust{i+1}' for i in range(15)],
    'Revenue': np.random.randint(1000, 10000, size=15)
}
df_segment = pd.DataFrame(data)

# Define revenue segments
conditions = [
    (df_segment['Revenue'] > 7000),
    (df_segment['Revenue'] > 4000) & (df_segment['Revenue'] <= 7000),
    (df_segment['Revenue'] <= 4000)
]
choices = ['High Value', 'Medium Value', 'Low Value']
df_segment['Segment'] = np.select(conditions, choices, default='Unknown')

# Aggregate revenue by segment
segment_summary = df_segment.groupby('Segment')['Revenue'].sum().reset_index()

# Set Seaborn theme
sns.set_theme(style="whitegrid")

# Plot segmentation as a bar chart
plt.figure(figsize=(12, 8))
sns.barplot(data=segment_summary, x='Segment', y='Revenue', palette='viridis')

# Enhance visualization
plt.title('Customer Segmentation by Revenue Contribution', fontsize=18, fontweight='bold')
plt.xlabel('Customer Segment', fontsize=14)
plt.ylabel('Total Revenue ($)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()

# Display detailed segmentation
print("Customer Segmentation Details:")
print(df_segment)
```

```python
# Dummy data: Monthly KPIs
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
revenue_growth = [5, 6, 7, 6.5, 8, 7.5]
profit_margin = [15, 16, 15.5, 16.2, 17, 16.8]

# Create DataFrame
df_kpi = pd.DataFrame({
    'Month': months,
    'Revenue Growth (%)': revenue_growth,
    'Profit Margin (%)': profit_margin
})

# Set Seaborn theme
sns.set_theme(style="whitegrid")

# Plot KPIs
plt.figure(figsize=(14, 8))
sns.lineplot(data=df_kpi, x='Month', y='Revenue Growth (%)', marker='o', label='Revenue Growth', linewidth=2.5)
sns.lineplot(data=df_kpi, x='Month', y='Profit Margin (%)', marker='o', label='Profit Margin', linewidth=2.5)

plt.title('KPI Monitoring: Revenue Growth and Profit Margin', fontsize=18, fontweight='bold')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Percentage (%)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import plotly.graph_objects as go

# Define nodes and links
nodes = ["Revenue", "COGS", "Gross Profit", "Operating Expenses", "EBITDA"]
links = [
    {"source": 0, "target": 1, "value": 400000},  # Revenue -> COGS
    {"source": 1, "target": 2, "value": 100000},  # COGS -> Gross Profit
    {"source": 2, "target": 3, "value": 50000},   # Gross Profit -> Operating Expenses
    {"source": 3, "target": 4, "value": 50000},   # Operating Expenses -> EBITDA
]

# Prepare data for the Sankey Diagram
fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=nodes,
    ),
    link=dict(
        source=[link["source"] for link in links],
        target=[link["target"] for link in links],
        value=[link["value"] for link in links],
    )
)])

fig.update_layout(title_text="EBITDA KPI Tree (Sankey Diagram)", font_size=14)
fig.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dummy data for EBITDA calculation
categories = ["Revenue", "COGS", "Gross Profit", "Operating Expenses", "EBITDA"]
values = [500000, -400000, 100000, -50000, 50000]

# Calculate cumulative values for the waterfall chart
cumulative = np.cumsum([0] + values[:-1])
cumulative_end = cumulative + values

# Plot the waterfall chart
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(categories, values, bottom=cumulative, color=['green', 'red', 'blue', 'orange', 'purple'])

# Annotate the bars
for bar, value, label in zip(bars, values, cumulative_end):
    ax.text(bar.get_x() + bar.get_width() / 2, label + (5000 if value > 0 else -5000),
            f"${value:,.0f}", ha='center', va='center', fontsize=10, color='black')

# Enhance visualization
ax.axhline(0, color='black', linewidth=1, linestyle='--')
plt.title('EBITDA Breakdown (Waterfall Chart)', fontsize=18, fontweight='bold')
plt.ylabel('Amount ($)', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Dummy data: EBITDA margin by segment
segments = ['Segment A', 'Segment B', 'Segment C', 'Segment D']
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
data = np.random.uniform(10, 30, size=(len(segments), len(quarters)))

# Create DataFrame
df_heatmap = pd.DataFrame(data, index=segments, columns=quarters)

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_heatmap, annot=True, fmt=".1f", cmap="coolwarm", linewidths=0.5)
plt.title('EBITDA Margin by Segment (Heatmap)', fontsize=18, fontweight='bold')
plt.xlabel('Quarter', fontsize=14)
plt.ylabel('Segment', fontsize=14)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import plotly.graph_objects as go

# Define nodes for the Sankey Diagram
nodes = [
    "Revenue", "Product Sales", "Service Revenue", "COGS", "Gross Profit",
    "Operating Expenses", "Salaries", "Marketing", "Other Expenses", "EBITDA"
]

# Define links and values
links = [
    {"source": 0, "target": 1, "value": 300000},  # Revenue -> Product Sales
    {"source": 0, "target": 2, "value": 200000},  # Revenue -> Service Revenue
    {"source": 1, "target": 3, "value": 250000},  # Product Sales -> COGS
    {"source": 2, "target": 3, "value": 100000},  # Service Revenue -> COGS
    {"source": 3, "target": 4, "value": 150000},  # COGS -> Gross Profit
    {"source": 4, "target": 5, "value": 50000},   # Gross Profit -> Operating Expenses
    {"source": 5, "target": 6, "value": 20000},   # Operating Expenses -> Salaries
    {"source": 5, "target": 7, "value": 15000},   # Operating Expenses -> Marketing
    {"source": 5, "target": 8, "value": 15000},   # Operating Expenses -> Other Expenses
    {"source": 4, "target": 9, "value": 100000}   # Gross Profit -> EBITDA
]

# Build Sankey diagram
fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=nodes,
    ),
    link=dict(
        source=[link["source"] for link in links],
        target=[link["target"] for link in links],
        value=[link["value"] for link in links],
    )
)])

# Add title and show
fig.update_layout(title_text="Detailed EBITDA Breakdown (Sankey Diagram)", font_size=14)
fig.show()
```

```python
import plotly.express as px
import pandas as pd

# Dummy data: Detailed breakdown of EBITDA components
data = {
    "Category": [
        "Revenue", "Revenue", "COGS", "COGS", "Operating Expenses", "Operating Expenses", "Operating Expenses", "EBITDA"
    ],
    "Sub-Category": [
        "Product Sales", "Service Revenue", "Direct Materials", "Direct Labor", "Salaries", "Marketing", "Other Expenses", "Net EBITDA"
    ],
    "Value": [
        500000, 200000, -300000, -100000, -50000, -20000, -30000, 200000
    ]
}

# Create DataFrame
df_treemap = pd.DataFrame(data)

# Build the treemap
fig = px.treemap(
    df_treemap,
    path=["Category", "Sub-Category"],
    values="Value",
    color="Value",
    color_continuous_scale="RdYlGn",
    title="EBITDA Components Treemap"
)

# Update layout for better readability
fig.update_layout(
    title_font_size=24,
    margin=dict(t=50, l=25, r=25, b=25)
)

# Show the plot
fig.show()
```
