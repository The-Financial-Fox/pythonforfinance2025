```python
# Import required libraries
import pandas as pd

# Load the Excel file
file_name = "/content/Dashboard in Excel.xlsx"  # Ensure this matches the file name you upload
sheet_name = "Dataset"  # Sheet name to be read

# Read the data from the specified sheet
data = pd.read_excel(file_name, sheet_name=sheet_name)

# Display the first few rows of the data
data.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate key metrics
total_sales = data['Sales'].sum()
total_profit = data['Profit'].sum()
average_profit_margin = (total_profit / total_sales) * 100  # As a percentage

# Create a simple display for the KPIs
fig, ax = plt.subplots(figsize=(12, 4))
fig.patch.set_facecolor('#f4f4f4')  # Set background color for the figure

# Define KPIs
kpis = [("Total Sales", f"£{total_sales:,.2f}"),
        ("Total Profit", f"£{total_profit:,.2f}"),
        ("Avg Profit Margin", f"{average_profit_margin:.2f}%")]

# Display KPIs in the Matplotlib figure
for i, (label, value) in enumerate(kpis):
    ax.text(0.2 + i * 0.25, 0.5, label, fontsize=16, fontweight='bold', color='black', ha='center')
    ax.text(0.2 + i * 0.25, 0.3, value, fontsize=14, color='#555555', ha='center')

# Remove axes and add title
ax.axis('off')
plt.title("Key Metrics", fontsize=18, fontweight='bold', color='black', pad=20)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare data for the Map Chart
sales_by_country = data.groupby('Country')['Sales'].sum().sort_values(ascending=False).reset_index()

# Prepare data for the Monthly Sales Bar Chart
monthly_sales = data.groupby('Month Name', sort=False)['Sales'].sum().reset_index()

# Sort months in calendar order
month_order = ["January", "February", "March", "April", "May", "June",
               "July", "August", "September", "October", "November", "December"]
monthly_sales['Month Name'] = pd.Categorical(monthly_sales['Month Name'], categories=month_order, ordered=True)
monthly_sales = monthly_sales.sort_values('Month Name')

# Create the dashboard layout
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle("Sales Analysis", fontsize=20, fontweight='bold', y=1.05)

# 1. Map Chart: Sales by Country
sns.barplot(x="Sales", y="Country", data=sales_by_country, ax=axes[0], palette="viridis")
axes[0].set_title("Sales by Country", fontsize=16, fontweight='bold')
axes[0].set_xlabel("Total Sales (£)", fontsize=12)
axes[0].set_ylabel("Country", fontsize=12)
axes[0].tick_params(axis='x', rotation=45)

# 2. Monthly Sales Bar Chart
sns.barplot(x="Month Name", y="Sales", data=monthly_sales, ax=axes[1], palette="coolwarm")
axes[1].set_title("Monthly Sales", fontsize=16, fontweight='bold')
axes[1].set_xlabel("Month", fontsize=12)
axes[1].set_ylabel("Total Sales (£)", fontsize=12)
axes[1].tick_params(axis='x', rotation=45)

# Adjust layout and show
plt.tight_layout()
plt.show()
```

```python
# Boxplot: Profit by Product
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Product', y='Profit', palette='muted')
plt.title("Profit Distribution by Product", fontsize=16, fontweight='bold')
plt.xlabel("Product", fontsize=12)
plt.ylabel("Profit (£)", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Pairplot: Key Metrics
# Select relevant numerical columns
pairplot_data = data[['Units Sold', 'Sales', 'COGS', 'Profit']]

# Create the pairplot
sns.pairplot(pairplot_data, kind="scatter", diag_kind="kde", corner=True, palette="coolwarm", height=2.5)
plt.suptitle("Pairplot of Key Metrics", y=1.02, fontsize=16, fontweight='bold')
plt.show()
```

```python
import numpy as np

# Prepare data for the Heatmap
heatmap_data = data.groupby(['Country', 'Segment'])['Sales'].sum().unstack()

# Heatmap: Sales by Country and Segment
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="YlGnBu", cbar_kws={'label': 'Total Sales (£)'})
plt.title("Sales by Country and Segment", fontsize=16, fontweight='bold')
plt.xlabel("Segment", fontsize=12)
plt.ylabel("Country", fontsize=12)
plt.tight_layout()
plt.show()

# Prepare data for the Dual-Axis Line Chart
time_series = data.groupby(['Month Name'], sort=False).agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

# Sort months in calendar order
time_series['Month Name'] = pd.Categorical(time_series['Month Name'], categories=month_order, ordered=True)
time_series = time_series.sort_values('Month Name')

# Dual-Axis Line Chart
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Sales on the first axis
ax1.plot(time_series['Month Name'], time_series['Sales'], color='blue', marker='o', label='Sales')
ax1.set_xlabel("Month", fontsize=12)
ax1.set_ylabel("Sales (£)", color='blue', fontsize=12)
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_title("Sales and Profit Over Time", fontsize=16, fontweight='bold')

# Plot Profit on the second axis
ax2 = ax1.twinx()  # Create a second y-axis
ax2.plot(time_series['Month Name'], time_series['Profit'], color='green', marker='o', label='Profit')
ax2.set_ylabel("Profit (£)", color='green', fontsize=12)
ax2.tick_params(axis='y', labelcolor='green')

# Add legends
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')

plt.tight_layout()
plt.show()
```

```python
!pip install python-pptx seaborn matplotlib

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from pptx import Presentation
from pptx.util import Inches
import os

# Load your dataset
file_name = "/content/Dashboard in Excel.xlsx"  # Replace with your file path
sheet_name = "Dataset"
data = pd.read_excel(file_name, sheet_name=sheet_name)

# Create a folder to save the figures
output_folder = "simple_visuals"
os.makedirs(output_folder, exist_ok=True)

# Function to add a slide with a title and an image
def add_slide_with_image(presentation, title, image_path):
    slide = presentation.slides.add_slide(presentation.slide_layouts[5])  # Blank slide layout
    slide.shapes.add_picture(image_path, Inches(1), Inches(1), width=Inches(8))  # Add image
    title_shape = slide.shapes.title
    if title_shape:
        title_shape.text = title

# Initialize the presentation
presentation = Presentation()

# 1. Bar Chart: Sales by Segment
sales_by_segment = data.groupby('Segment')['Sales'].sum().reset_index()
bar_chart_path = os.path.join(output_folder, "sales_by_segment.png")
plt.figure(figsize=(8, 6))
sns.barplot(x="Segment", y="Sales", data=sales_by_segment, palette="viridis")
plt.title("Sales by Segment", fontsize=16)
plt.xlabel("Segment")
plt.ylabel("Sales (£)")
plt.savefig(bar_chart_path)
plt.close()
add_slide_with_image(presentation, "Sales by Segment", bar_chart_path)

# 2. Line Chart: Sales Over Time (Month)
monthly_sales = data.groupby('Month Name', sort=False)['Sales'].sum().reset_index()
month_order = ["January", "February", "March", "April", "May", "June",
               "July", "August", "September", "October", "November", "December"]
monthly_sales['Month Name'] = pd.Categorical(monthly_sales['Month Name'], categories=month_order, ordered=True)
monthly_sales = monthly_sales.sort_values('Month Name')

line_chart_path = os.path.join(output_folder, "sales_over_time.png")
plt.figure(figsize=(8, 6))
sns.lineplot(x="Month Name", y="Sales", data=monthly_sales, marker='o', color='blue')
plt.title("Sales Over Time", fontsize=16)
plt.xlabel("Month")
plt.ylabel("Sales (£)")
plt.xticks(rotation=45)
plt.savefig(line_chart_path)
plt.close()
add_slide_with_image(presentation, "Sales Over Time", line_chart_path)

# 3. Boxplot: Profit by Segment
boxplot_path = os.path.join(output_folder, "profit_by_segment.png")
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x="Segment", y="Profit", palette="Set2")
plt.title("Profit Distribution by Segment", fontsize=16)
plt.xlabel("Segment")
plt.ylabel("Profit (£)")
plt.savefig(boxplot_path)
plt.close()
add_slide_with_image(presentation, "Profit Distribution by Segment", boxplot_path)

# Save the PowerPoint file
output_pptx = "Simple_Visuals_Presentation.pptx"
presentation.save(output_pptx)

print(f"PowerPoint presentation saved as {output_pptx}")
```
